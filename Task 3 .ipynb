# Exploratory Data Analysis:

# Univariate Anaylsis:
Univariate analysis explores each variable in a data set, separately. It looks at the range of values, as well as the central tendency of the values. It describes the pattern of response to the variable. It describes each variable on its own.

```python
plt.figure(figsize=(8,8))
plt.title('Ship-Mode Analysis')
sns.countplot(x=s_store['Ship Mode'], edgecolor = 'b', order = s_store['Ship Mode'].value_counts().index)
plt.show()
```

* From the analysis of 'Ship Mode', we concluded that, 'Standard Class' got the high no. of entries than any other mode.
* Then after,Second Class, First Class and the last one is Same Day.

```python
plt.figure(figsize=(8,8))
plt.title('Segment Analysis')
sns.countplot(x=s_store['Segment'], edgecolor= 'b', order=s_store['Segment'].value_counts().index)
plt.show()
```

* From the analysis of 'Segment', we concluded that, 'Consumer' got the high no. of entries than any other mode.
* Then after,Corporate Segment and the last one is Home Office.

```python
plt.figure(figsize=(12,10))
plt.title('City Analysis')
sns.countplot(x= s_store['City'], edgecolor ='b', order=s_store['City'].value_counts().iloc[:10].index)
plt.show()
```

* From the above analysis we concluded that, 'New York' city is on the high demand for SuperStore. Secondly, 'Los Angles' and 'Philadelphia'.

```python
plt.figure(figsize=(15,15))
plt.title('State Analysis')
plt.xticks(rotation= 'vertical')
sns.countplot(x= s_store['State'], edgecolor ='b', order=s_store['State'].value_counts().index)
plt.show()
```

* From the above state analysis we got conlcuded that our busines are on high demand in the state named California, New York and Texas.

```python
plt.figure(figsize=(12,8))
plt.title('Postal Code Analysis')
sns.countplot(x= s_store['Postal Code'], edgecolor ='b', order=s_store['Postal Code'].value_counts().iloc[:10].index)
plt.show()
```

* From the above postal code analysis we concluded that, the pin code 10035, 10024 and 10009 are more active towards SuperStore.

```python
plt.figure(figsize=(8,8))
plt.title('Region-wise Analysis')
sns.countplot(x=s_store['Region'], edgecolor='b', order=s_store['Region'].value_counts().index)
plt.show()
```

* From the above analysis we got concluded that, West is high on the demand by SuperStore. Secondly, East , Central south.

```python
plt.figure(figsize=(8,8))
plt.title('Category-wise Analysis')
sns.countplot(x=s_store['Category'], edgecolor='b', order=s_store['Category'].value_counts().index)
plt.show()
```

* From the above analysis OFfice supplyies are on high demand from the SuperStore.

```python
plt.figure(figsize=(15,8))
plt.title('Sub-Category-wise Analysis')
sns.countplot(x=s_store['Sub-Category'], edgecolor='b', order=s_store['Sub-Category'].value_counts().index)
plt.show()
```

* From the above analysis Binders, Papers and Furnishing are high demand.

# Bivariate Analysis:
Bivariate analysis is one of the simplest forms of quantitative (statistical) analysis. It involves the analysis of two variables (often denoted as X, Y), for the purpose of determining the empirical relationship between them. Bivariate analysis can be helpful in testing simple hypotheses of association.

```python
plt.figure(figsize=(15,15))
sns.countplot(x='Sub-Category', hue='Category', data=s_store, edgecolor='black', order=s_store['Sub-Category'].value_counts().index)
plt.xticks(rotation='vertical')
plt.show()
```

* From the above bivariate analysis, Binders from Office supplies are on high demand, followed by papers.

```python
plt.figure(figsize=(15,15))
sns.factorplot(x= 'Ship Mode', y='Profit', hue='Category', kind='bar', data= s_store)
plt.xticks(rotation='vertical')
plt.show()
```
