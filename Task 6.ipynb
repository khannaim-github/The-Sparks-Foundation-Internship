# Data sampling:

Perform some data sampling. In which divide our given data set into train and test set, where train data set is used to train our given data set and test data is used to test our trained data set. 

* Train and Test data set is divided into 80/20 ratio.
* In this spliting, all the variables are considered as independant variables except Species, which is considered as dependant variable or target variable.

```python
from sklearn.model_selection import train_test_split
iris_dt_train, iris_dt_test = train_test_split(iris_dt,test_size=0.2,random_state=123 )
```

```python
iris_dt_train_x = iris_dt_train.iloc[:, 0:4]
iris_dt_train_y = iris_dt_train.iloc[:, -1]
```

```python
iris_dt_test_x = iris_dt_test.iloc[:, 0:4]
iris_dt_test_y = iris_dt_test.iloc[:, -1]
```

```python
print('Shape of iris_dt_train_x', iris_dt_train_x.shape)
print('Shape of iris_dt_train_y', iris_dt_train_y.shape)
print('Shape of iris_dt_test_x', iris_dt_test_x.shape)
print('Shape of iris_dt_test_y',iris_dt_test_y.shape)
```

* Here we can check the shape of train and test data set, including dependant and independant variable.

# Model building:

Let's perform some model building over the trained data set using DecisionTree Classifier which is imported from sklearn.tree library.

Apply the parameter 'class_weight='balanced''to  balance the confusion matrix by avoiding the overfitting and underfitting.

```python
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(class_weight='balanced')
```

* Let's fit the built model, over the train data.

```python
dt_fit= dt.fit(iris_dt_train_x, iris_dt_train_y)
```

```python
pred_dt = dt.predict(iris_dt_test_x)
```

* Let's have some confusion matrix to check the accuracy of the formed data set.

```python
tab_dt = confusion_matrix(pred_dt, iris_dt_test_y)
tab_dt
```

* Let's check the accuracy of the formed model.

```python
Acc_dt = tab_dt.diagonal().sum() / tab_dt.sum() *100
Acc_dt
```

* Based on the accuracy of the given model and confusion matrix, we can conclude that our model built is perfectly built.
